```python
def shuffle_waveform(audio, sec=1.0):
    chunk = int(sr*sec)
    step = len(audio)//chunk
    newaudio = np.array([0])
    np.random.seed(0)
    for idx in np.random.choice(np.arange(step), step, replace=False):
        start = idx*chunk
        newaudio = np.append(newaudio, audio[start:start+chunk])
    return newaudio[1:]
```
